```python
# lib import 
import matplotlib.pyplot as plt
import nltk
#nltk.download() (!pip install nltk)
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
#conda install -c conda-forge wordcloud
import string
```

```python
# creating corpus using file
corpus = PlaintextCorpusReader('Files.zip', '.*', encoding = "ISO-8859-1")
```

```python
# read form disk, shows 100 firts files 
files = corpus.fileids()
#primeiro arquivo
files[0]
```

```python
# Access first text file 
text = corpus.raw('Files/1.txt')
text
```

```python
# access all words in all files
all_text = corpus.raw()
all_text
```

```python
# getting all the words and couting
words = corpus.words()
#accessing by index
words[170]
```

```python
#quantity
len(words)
```

```python
# Using NLTK we get the stopwords in english Usando o NLTK, obtemos as stop word em ingls
stops = stopwords.words('english')
#stops = stopwords.words('portuguese')
stops
```

```python
# Defining colors at the cloudword
Colors_map = ListedColormap(['orange', 'green', 'red', 'magenta'])
# creating clouadword using max of 100 words and stopwords
cloud = WordCloud(background_color = 'white',
                  colormap = Colors_map,
                  stopwords = stops,
                  max_words = 100)
# creation and visualization of the cloud
cloud.generate(all_text)
plt.imshow(cloud)
```

```python
# creating a new word list without stopwords 
stopwords_words = [p for p in words if p not in stops]
len(stopwords_words)
```

```python
# removing ponctuation and creating a new cloudword without ponctuation and stopwords
Without_ponctuation_words = [p for p in stopwords_words if p not in string.punctuation]
len(Without_ponctuation_words)
```

```python
# frequency nd most common words
frequency = nltk.FreqDist(Without_ponctuation_words)
frequency
```

```python
#most common
most_common = frequency.most_common(100)
most_common
```

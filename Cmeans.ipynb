```python
# lib import 
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
#pip install -U scikit-fuzzy (executar no Anaconda Prompt)
```

```python
# Loading  database 
iris = datasets.load_iris()
```

```python
# appling 3 clusters using the transposed matrix other parameters as default
r = skfuzzy.cmeans(data = iris.data.T, c = 3, m = 2, error = 0.005,
                   maxiter = 1000, init = None)
```

```python
# percentage of a register inside a cluster 
prevision_percentage = r[1]
```

```python
# probability of a register to belong at least a one cluster 
for x in range(150):
  print( prevision_percentage[0][x] ,prevision_percentage[1][x] ,prevision_percentage[2][x] )
```

```python
# contingency matrix to compare with the original one 
prevision = prevision_percentage.argmax(axis = 0)
results = confusion_matrix(iris.target, prevision)
results
```

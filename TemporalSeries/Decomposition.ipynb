```python
# lib import 
import pandas as pd
import matplotlib.pylab as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime
#convert registers for matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Loading  dataset, changing data string -> data and creating temporal series
base = pd.read_csv('AirPassengers.csv')
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
base = pd.read_csv('AirPassengers.csv', parse_dates = ['Month'],
                   index_col = 'Month', date_parser = dateparse)
ts = base['#Passengers']
```

```python
# temporal series visualization
plt.plot(ts)
```

```python
# temporal serie decomposition creating a variable for each format
decomposition = seasonal_decompose(ts)
```

```python
#trend
trend = decomposition.trend
trend
```

```python
#seasonal
seasonal = decomposition.seasonal
seasonal
```

```python
#random
random= decomposition.resid
random
```

```python
# one chart for each format
plt.plot(seasonal)
```

```python
plt.plot(trend)
```

```python
plt.plot(random)
```

```python
plt.subplot(4,1,1)
plt.plot(ts, label = 'Original')
plt.legend(loc = 'best')

# just trend
plt.subplot(4,1,2)
plt.plot(trend, label = 'Trend')
plt.legend(loc = 'best')

# just seasonal
plt.subplot(4,1,3)
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc = 'best')

# just random
plt.subplot(4,1,4)
plt.plot(random, label = 'Random')
plt.legend(loc = 'best')
plt.tight_layout()
```

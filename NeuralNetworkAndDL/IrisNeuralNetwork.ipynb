```python
# lib import 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
```

```python
# Loading database and creating previsors 
base = datasets.load_iris()
previsores = base.data
vclass = base.target
vclass
```

```python
# class to dummy because the neural network have 3 exit layers 
vclass_dummy = np_utils.to_categorical(vclass)
vclass_dummy
```

```python
# 70/30 split 
X_training, X_test, y_training, y_test = train_test_split(previsores,
                                                                  vclass_dummy,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```

```python
# creating neural network with Sequential class 
model = Sequential()
#first hidden layer
model.add(Dense(units = 5, input_dim = 4))
#second hidden layer 
model.add(Dense(units = 4))
# softmax because we have more than 2 vclass
# generate a probability in each neuron
model.add(Dense(units = 3, activation = 'softmax'))
```

```python
# neural network structure 
model.summary()
```

```python
# parameter config 
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
               metrics = ['accuracy'])
# traning and deviding to validate 
model.fit(X_training, y_training, epochs = 1000,
           validation_data = (X_test, y_test))
```

```python
# predicting and changing variable to true and false according to threshold
prediction = model.predict(X_test)
prediction = (prediction > 0.5)
prediction
```

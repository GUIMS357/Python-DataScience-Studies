```python
# libimport 
from igraph import Graph
from igraph import plot
```

```python
# creating a graph with weights
graph5 = Graph(edges = [(0,1),(2,3),(0,2),(0,3)], directed = True)
graph5.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio']
graph5.vs['weight'] = [40,30,30,25]
graph5.es['friendshiptype'] = ['friend', 'enemy', 'enemy', 'friend']
graph5.es['weight'] = [1,2,1,3]
```

```python
# visualizing information in the vertexes
for v in graph5.vs:
    print(v)
```

```python
# visualizing information in the edges 
for e in graph5.es:
    print(e)
```

```python
# Devining color to vertexes 
graph5.vs['color'] = ['blue', 'red', 'yellow', 'green']
plot(graph5, bbox=(0,0,300,300),vertex_color = graph5.vs['color'])
```

```python
# weight to edges
plot(graph5, bbox=(300,300),  edge_width = graph5.es['weight'],
     vertex_color = graph5.vs['color'])
```

```python
# weight to vertexes 
plot(graph5, bbox=(300,300), vertex_size = graph5.vs['weight'],
     edge_width = graph5.es['weight'],
     vertex_color = graph5.vs['color'])
```

```python
# curvature
plot(graph5, bbox=(300,300), vertex_size = graph5.vs['weight'],
     edge_width = graph5.es['weight'],
     vertex_color = graph5.vs['color'],
     edge_curved = 0.4)
```

```python
# Form
plot(graph5, bbox=(0,0,300,300), vertex_size = graph5.vs['weight'],
     edge_width = graph5.es['weight'],
     vertex_color = graph5.vs['color'],
     edge_curved = 0.4, vertex_shape = 'triangle')
```

```python
# lib import
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# creating a standard deviation variable using rvs(100 elements)
data = norm.rvs(size = 1000)
data
```

```python
#histogram
plt.hist(data, bins = 20)
plt.title('data')
```

```python
# generating chart to verify if it is a normal distribution
fig, ax = plt.subplots()
stats.probplot(data, fit=True,   plot=ax)
plt.show()
```

```python
# shapiro test
#second argument is p value
stats.shapiro(data)
```

Dados não normais

```python
data2 = skewnorm.rvs(4, size=1000)
```

```python
#histograma
plt.hist(data2, bins = 20)
plt.title('data')
```

```python
# generating chart to verify if it is a normal distribution
fig, ax = plt.subplots()
stats.probplot(data2, fit=True,   plot=ax)
plt.show()
```

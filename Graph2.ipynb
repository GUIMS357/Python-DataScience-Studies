```python
# lib miport
from igraph import Graph
from igraph import plot
```

```python
# directed graph
graph1 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = True)
graph1.vs['label'] = range(graph1.vcount())
print(graph1)
```

```python
# creating a non directed 
graph2 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = False)
graph2.vs['label'] = range(graph2.vcount())
print(graph2)
```

```python
#chart 
plot(graph2, bbox=(0,0,300,300))
```

```python
# adding vertex and edges
graph3 = Graph(directed = False)
graph3.add_vertices(10)
#graph3.add_vertex(16)
graph3.add_edges([(0,1),(2,2),(2,3),(3,0)])
graph3.vs['label'] = range(graph3.vcount())
print(graph3)
plot(graph3, bbox=(0,0,300,300))
```

```python
# creating 4th graph and adding custom labels
graph4 = Graph(directed = False)
graph4.add_vertices(5)
graph4.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
graph4.add_vertex(5)
graph4.add_vertex(6)
graph4.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
plot(graph4, bbox=(0,0,300,300))
```
